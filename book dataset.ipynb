```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('final_book_dataset_kaggle.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df.replace('Nan', np.nan, inplace = True)
```

```python
df.dropna(axis = 0, inplace = True)
```

```python
df.columns
```

```python
pythonbooks = df[df['title'].str.contains("Python")]

bestpythonbooks = pythonbooks.nlargest(7, ['n_reviews','avg_reviews'])
bestpythonbooks
```

```python
mlbooks = df[df['title'].str.contains("Machine Learning")]

bestmlbooks = mlbooks.nlargest(7, ['n_reviews','avg_reviews'])
bestmlbooks
```

```python
maxreviews = df['n_reviews'].max() 
reviewbook = df.loc[df['n_reviews'] == maxreviews, 'title'].iloc[0]
print(reviewbook, 'is the most reviewed book with max ', maxreviews, 'reviews')
```

```python
plt.figure(figsize=(20,20))
sns.barplot(x='n_reviews',y='title', data = df.head(20))
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 6)
plt.ylabel('Book Title')
plt.xlabel('Number of Reviews')
plt.title('Number of reviews per book')
```

```python
fivestar = df['star5'].max()
fivestarbook = df.loc[df['star5'] == fivestar, 'title'].iloc[0]
print(fivestarbook, ' got most percentage of 5 stars (', fivestar, ')')
```

```python
n_reviews_fivestarbook = df.loc[df['title'] == fivestarbook, 'n_reviews'].iloc[0]
print(fivestarbook, 'got ',n_reviews_fivestarbook, ' reviews out of which 5 stars were ', fivestar )
```

```python
maximum_reviews = df['n_reviews'].max()
fivestar_on_max_n_reviewsbook = df.loc[df['title'] == reviewbook, 'star5'].iloc[0]
print(reviewbook, 'got', fivestar_on_max_n_reviewsbook, 'percent of 5 stars and ', maximum_reviews, 'reviews')
```

```python
plt.figure(figsize=(20,20))
sns.barplot(x='star5',y='title', data = df.head(50))
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.yticks(fontsize= 6)
plt.ylabel('Book Title')
plt.xlabel('Rating')
plt.title('Books with 5 star rating')
```

```python
most_reviewed = df['n_reviews'].max()
star_publisher = df.loc[df['n_reviews'] == most_reviewed, 'publisher'].iloc[0]

print(star_publisher, ' got maximum number of reviews (', most_reviewed, ')')
```

```python
plt.figure(figsize=(15,8))
sns.catplot(x = 'n_reviews', y = 'publisher',data = df.head(50), kind = 'swarm')
plt.yticks(fontsize= 6)
plt.ylabel('Publisher')
plt.xlabel('number of reviews')
plt.title('Publishers and number of reviews per their books')
```

```python
most_reviewed = df['n_reviews'].max()
star_author = df.loc[df['n_reviews'] == most_reviewed, 'author'].iloc[0]

print(star_author, ' got maximum number of reviews (', most_reviewed, ')')
```

```python
plt.figure(figsize=(100, 80))
sns.catplot(x = 'n_reviews', y = 'author',data = df.head(20), kind = 'bar')
plt.yticks(fontsize= 6)
plt.ylabel('author')
plt.xlabel('number of reviews')
plt.title('Authors and number of reviews per their books')
```

```python
cost = df['price'].max()
cost_book = df.loc[df['price'] == cost, 'title'].iloc[0]

print(cost_book, 'is the costliest book priced at $', cost)
```

```python
least_cost = df['price'].min()
least_costliest_book = df.loc[df['price'] == least_cost, 'title'].iloc[0]

print(least_costliest_book, 'is the costliest book priced at $', least_cost)
```

```python
plt.figure(figsize=(15,8))
sns.catplot(x = 'price', y = 'title',data = df.head(20), kind = 'swarm')
plt.yticks(fontsize= 6)
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.ylabel('title')
plt.xlabel('price in $')
plt.title('Title and its price')
```

```python
plt.figure(figsize=(15,8))

df['avg_reviews'].plot()
plt.xlabel('count')
plt.ylabel('reviews')
plt.show()
```
